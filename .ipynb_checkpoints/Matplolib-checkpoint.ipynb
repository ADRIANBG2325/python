se sigue el mismo procedimiento para pintar una funcion matematica.

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2

plt.plot(x,y)
plt.show()
```

tambien puede modificarse el estilo de la grafica para que contenga mas informacin

```python
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("Chido")
plt.label("y = x\^2")
plt.grid(True)
plt.show()
```

Pueden superponerse graficas y cambiar el estilo de las lineas

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x+1
plt.plot(x,y,'b--',x,y2,'g')
plt.show()
```

```python
# separando en diferentes lineas las funciones.
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x+1
plt.title("Square Function")
plt.plot(x,y,'b--')
plt.plot(x,y2,'g')
plt.xlabel("Chido")
plt.label("y = x^2")
plt.grid(True)
plt.show()
```

para poder diferencial entre ambas funciones siempre es recomendable añadir una leyenda

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x+1
plt.title("Square Function")
plt.plot(x,y,'b--',label="x^2")
plt.plot(x,y2,'g',label="x+1")
plt.xlabel("Chido")
plt.label("y = x^2")
plt.grid(True)
plt.legend(loc ="best")
#lo situa en la mejor ubicacion
plt.show()
```

tambien puede crearse dos graficas que no se super pongan estas graficas se organizan en un grid  se denominan sub plots

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x+1
plt.title("Square Function")
plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.subplot(1,2,2)
plt.plot(x,y2,'g')
plt.xlabel("Chido")
plt.label("y = x^2")
plt.grid(True)

#lo situa en la mejor ubicacion
plt.show()
```

para q la graficas no quede tan ajustadas se puede hacer la figura mas grande

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x+1

plt.figure(figsize=(14,6))

plt.title("Square Function")
plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.xlabel("x1",fontsize =14)
plt.xlabel("y1",fontsize =14)

plt.subplot(1,2,2)
plt.plot(x,y2,'g')
plt.xlabel("x2",fontsize =14)
plt.xlabel("y2",fontsize =14)

plt.show()
```

## scatter plots

```python
from numpy.random import rand
x,y=rand(2,100)
plt.scatter(x,y)
plt.show()
```

```python
from numpy.random import rand
x,y=rand(2,100)
x2,y2=rand(2,100)
plt.scatter(x,y,c='red')
plt.scatter(x2,y2,c='blue')
plt.title("scatter")
plt.grid(True)
plt.show()
```

histograma

```python
data=[1,1.1,1.8,2,2.1,3.2,3,3,3,3]
plt.subplot(211)# 2 row, 1 col, 1 sb
plt.hist(data, bins=10, rwidth=0.8)

plt.subplot(212)# 2 row, 1 col, 1 sb
plt.hist(data, bins=[1,1.5,2,2.5,3], rwidth=0.95)
plt.xlabel('Value')
plt.ylabel('frecquency')
plt.show()
```
